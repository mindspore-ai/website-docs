# дҪҝз”ЁеҮҪж•°еҸҳжҚўи®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳөе’Ңй»‘еЎһзҹ©йҳө


## йӣ…еҸҜжҜ”зҹ©йҳө

еңЁд»Ӣз»ҚMindSporeжҸҗдҫӣзҡ„и®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳөзҡ„ж–№жі•д№ӢеүҚпјҢйҰ–е…ҲеҜ№йӣ…еҸҜжҜ”зҹ©йҳөиҝӣиЎҢд»Ӣз»ҚгҖӮ

жҲ‘д»¬йҰ–е…Ҳе®ҡд№үдёҖдёӘжҳ е°„$\textbf{f}$:

$$R^{n} \longrightarrow R^{m}$$

жҲ‘д»¬еңЁиҝҷйҮҢдҪҝз”Ёз¬ҰеҸ·$\longmapsto$иЎЁзӨәйӣҶеҗҲе…ғзҙ д№Ӣй—ҙзҡ„жҳ е°„пјҢе№¶еҠ зІ—жүҖжңүд»ЈиЎЁеҗ‘йҮҸзҡ„з¬ҰеҸ·пјҢеӣ жӯӨжңүпјҡ

$$\textbf{x} \longmapsto \textbf{f}(\textbf{x})$$

е…¶дёӯ$\textbf{x} = (x_{1}, x_{2},\dots, x_{n})$пјҢ$\textbf{f(x)} = (f_{1}(\textbf{x}), f_{2}(\textbf{x}),\dots,f_{m}(\textbf{x}))$гҖӮ

жҲ‘д»¬е°ҶжүҖжңүд»Һ$R^{n}$еҲ°$R^{m}$зҡ„жҳ е°„жһ„жҲҗзҡ„йӣҶеҗҲи®°дёә$F_{n}^{m}$гҖӮ
еңЁиҝҷйҮҢпјҢжҲ‘д»¬е°Ҷд»ҺдёҖдёӘеҮҪж•°пјҲжҳ е°„пјүйӣҶеҲ°еҸҰдёҖдёӘеҮҪж•°пјҲжҳ е°„пјүйӣҶзҡ„жҳ е°„з§°дёәдёҖдёӘж“ҚдҪңгҖӮжҳ“еҫ—пјҢжўҜеәҰж“ҚдҪң$\nabla$жҳҜеңЁеҮҪж•°йӣҶ$F_{n}^{1}$дёҠеҗҢж—¶иҝӣиЎҢnж¬ЎеҒҸеҜјж•°зҡ„ж“ҚдҪңпјҢе°Ҷе…¶е®ҡд№үдёәдёҖдёӘд»ҺеҮҪж•°йӣҶ$F_{n}^{1}$еҲ°$F_{n}^{n}$зҡ„жҳ е°„пјҡ

$$\nablaпјҡF_{n}^{1} \longrightarrow F_{n}^{n}$$

е№ҝд№үзҡ„жўҜеәҰж“ҚдҪң$\partial$пјҢиў«е®ҡд№үдёәеңЁеҮҪж•°йӣҶ$F_{n}^{m}$дёҠеҗҢж—¶иҝӣиЎҢnж¬ЎеҒҸеҜјж•°зҡ„ж“ҚдҪңпјҢ

$$\partial: F_{n}^{m} \longrightarrow F_{n}^{m \times n}$$

йӣ…еҸҜжҜ”зҹ©йҳөе°ұжҳҜе°Ҷж“ҚдҪң$\partial$дҪңз”ЁдәҺ$\textbf{f}$еҗҺеҫ—еҲ°зҡ„з»“жһңпјҢеҚі

$$\textbf{f} \longmapsto \partial \textbf{f} = (\frac{\partial \textbf{f}}{\partial x_{1}}, \frac{\partial \textbf{f}}{\partial x_{2}}, \dots, \frac{\partial \textbf{f}}{\partial x_{n}})$$

еҫ—еҲ°йӣ…еҸҜжҜ”зҹ©йҳөпјҢ

$$J_{f} = \begin{bmatrix}
  \frac{\partial f_{1}}{\partial x_{1}} &\frac{\partial f_{1}}{\partial x_{2}} &\dots &\frac{\partial f_{1}}{\partial x_{n}}  \\
  \frac{\partial f_{2}}{\partial x_{1}} &\frac{\partial f_{2}}{\partial x_{2}} &\dots &\frac{\partial f_{2}}{\partial x_{n}} \\
  \vdots &\vdots &\ddots &\vdots \\
  \frac{\partial f_{m}}{\partial x_{1}} &\frac{\partial f_{m}}{\partial x_{2}} &\dots &\frac{\partial f_{m}}{\partial x_{n}}
\end{bmatrix}$$

йӣ…еҸҜжҜ”зҹ©йҳөзҡ„еә”з”ЁпјҡеңЁиҮӘеҠЁеҫ®еҲҶзҡ„еүҚеҗ‘жЁЎејҸдёӯпјҢжҜҸдёҖж¬ЎеүҚеҗ‘дј ж’ӯпјҢеҸҜд»ҘжұӮеҮәйӣ…еҸҜжҜ”зҹ©йҳөзҡ„дёҖеҲ—гҖӮеңЁиҮӘеҠЁеҫ®еҲҶзҡ„еҸҚеҗ‘жЁЎејҸдёӯпјҢжҜҸдёҖж¬ЎеҸҚеҗ‘дј ж’ӯпјҢжҲ‘д»¬еҸҜд»Ҙи®Ўз®—еҮәйӣ…еҸҜжҜ”зҹ©йҳөзҡ„дёҖиЎҢгҖӮ


## и®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳө

дҪҝз”Ёж ҮеҮҶзҡ„иҮӘеҠЁеҫ®еҲҶзі»з»ҹеҫҲйҡҫй«ҳж•Ҳең°и®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳөпјҢдҪҶMindSporeеҚҙжҸҗдҫӣдәҶиғҪеӨҹй«ҳж•Ҳи®Ўз®—йӣ…еҸҜжҜ”зҡ„ж–№жі•пјҢдёӢйқўеҜ№иҝҷдәӣж–№жі•иҝӣиЎҢд»Ӣз»ҚгҖӮ

йҰ–е…ҲпјҢжҲ‘д»¬е®ҡд№үдёҖдёӘеҮҪж•°`forecast`пјҢиҜҘеҮҪж•°жҳҜдёҖдёӘз®ҖеҚ•зҡ„зәҝжҖ§еҮҪж•°пјҢе№¶еёҰжңүдёҖдёӘйқһзәҝжҖ§жҝҖжҙ»еҮҪж•°гҖӮ

```python
import time
import mindspore
from mindspore import ops
from mindspore import jacrev, jacfwd, vmap, vjp, jvp, grad
import numpy as np

mindspore.set_seed(1)

def forecast(weight, bias, x):
    return ops.dense(x, weight, bias).tanh()
```

жҺҘдёӢжқҘпјҢжҲ‘д»¬жһ„йҖ дёҖдәӣж•°жҚ®пјҡдёҖдёӘжқғйҮҚеј йҮҸ`weight`пјҢдёҖдёӘеҒҸе·®еј йҮҸ`bias`пјҢиҝҳжңүдёҖдёӘиҫ“е…Ҙеҗ‘йҮҸ`x`гҖӮ

```python
D = 16
weight = ops.randn(D, D)
bias = ops.randn(D)
x = ops.randn(D)
```

еҮҪж•°`forecast`еҜ№иҫ“е…Ҙеҗ‘йҮҸ`x`еҒҡеҰӮдёӢжҳ е°„еҸҳжҚўпјҢ$R^{D}\overset{}{\rightarrow}R^{D}$гҖӮ
MindSporeеңЁиҮӘеҠЁеҫ®еҲҶзҡ„иҝҮзЁӢдёӯпјҢдјҡи®Ўз®—еҗ‘йҮҸ-йӣ…еҸҜжҜ”з§ҜгҖӮдёәдәҶи®Ўз®—жҳ е°„$R^{D}\overset{}{\rightarrow}R^{D}$зҡ„е®Ңж•ҙйӣ…еҸҜжҜ”зҹ©йҳөпјҢжҲ‘д»¬жҜҸж¬ЎдҪҝз”ЁдёҚеҗҢзҡ„еҚ•дҪҚеҗ‘йҮҸйҖҗиЎҢжқҘи®Ўз®—е®ғгҖӮ

```python
def partial_forecast(x):
    return ops.dense(x, weight, bias).tanh()

_, vjp_fn = vjp(partial_forecast, x)

def compute_jac_matrix(unit_vectors):
    jacobian_rows = [vjp_fn(vec)[0] for vec in unit_vectors]
    return ops.stack(jacobian_rows)


unit_vectors = ops.eye(D)
jacobian = compute_jac_matrix(unit_vectors)
print(jacobian.shape)
print(jacobian[0])
```

еңЁ`compute_jac_matrix`дёӯпјҢдҪҝз”ЁforеҫӘзҺҜйҖҗиЎҢи®Ўз®—зҡ„ж–№ејҸи®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳөпјҢи®Ўз®—ж•ҲзҺҮе№¶дёҚй«ҳгҖӮMindSporeжҸҗдҫӣ`jacrev`жқҘи®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳөпјҢ`jacrev`зҡ„е®һзҺ°еҲ©з”ЁдәҶ`vmap`пјҢ`vmap`еҸҜд»Ҙж¶ҲйҷӨ`compute_jac_matrix`дёӯзҡ„forеҫӘзҺҜе№¶еҗ‘йҮҸеҢ–ж•ҙдёӘи®Ўз®—иҝҮзЁӢгҖӮ`jacrev`зҡ„еҸӮж•°`grad_position`жҢҮе®ҡи®Ўз®—иҫ“еҮәзӣёеҜ№дәҺе“ӘдёӘеҸӮж•°зҡ„йӣ…еҸҜжҜ”зҹ©йҳөгҖӮ


```python
from mindspore import jacrev
jacrev_jacobian = jacrev(forecast, grad_position=2)(weight, bias, x)
assert np.allclose(jacrev_jacobian.asnumpy(), jacobian.asnumpy())
```

жҺҘдёӢжқҘеҜ№`compute_jac_matrix`е’Ң`jacrev`зҡ„жҖ§иғҪиҝӣиЎҢеҜ№жҜ”гҖӮйҖҡеёёжғ…еҶөдёӢпјҢ`jacrev`зҡ„жҖ§иғҪжӣҙеҘҪпјҢеӣ дёәдҪҝз”Ё`vmap`иҝӣиЎҢеҗ‘йҮҸеҢ–иҝҗз®—пјҢдјҡжӣҙе……еҲҶең°еҲ©з”ЁзЎ¬д»¶пјҢеҗҢж—¶и®Ўз®—еӨҡз»„ж•°жҚ®пјҢйҷҚдҪҺи®Ўз®—ејҖй”ҖпјҢд»ҘиҺ·еҫ—жӣҙеҘҪзҡ„жҖ§иғҪгҖӮ

и®©жҲ‘д»¬зј–еҶҷдёҖдёӘеҮҪж•°пјҢеңЁеҫ®з§’йҮҸзә§дёҠиҜ„дј°дёӨз§Қж–№жі•зҡ„жҖ§иғҪгҖӮ

```python
def perf_compution(func, run_times, *args, **kwargs):
    start_time = time.perf_counter()
    for _ in range(run_times):
        func(*args, **kwargs)
    end_time = time.perf_counter()
    cost_time = (end_time - start_time) * 1000000
    return cost_time


run_times = 500
xp = x.copy()
compute_jac_matrix_cost_time = perf_compution(compute_jac_matrix, run_times, xp)
jac_fn = jacrev(forecast, grad_position=2)
jacrev_cost_time = perf_compution(jac_fn, run_times, weight, bias, x)
print(f"compute_jac_matrix run {run_times} times, cost time {compute_jac_matrix_cost_time} microseconds.")
print(f"jacrev run {run_times} times, cost time {jacrev_cost_time} microseconds.")
```

еҲҶеҲ«иҝҗиЎҢ`compute_jac_matrix`е’Ң`jacrev`500ж¬ЎпјҢз»ҹи®Ўе®ғд»¬ж¶ҲиҖ—зҡ„ж—¶й—ҙгҖӮ

дёӢйқўи®Ўз®—пјҢзӣёиҫғдәҺдҪҝз”Ё`compute_jac_matrix`пјҢдҪҝз”Ё`jacrev`и®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳөпјҢжҖ§иғҪжҸҗеҚҮзҡ„зҷҫеҲҶжҜ”гҖӮ

```python
def perf_cmp(first, first_descriptor, second, second_descriptor):
    faster = second
    slower = first
    gain = (slower - faster) / slower
    if gain < 0:
        gain *= -1
    final_gain = gain*100
    print(f" Performance delta: {final_gain:.4f} percent improvement with {second_descriptor}. ")

perf_cmp(compute_jac_matrix_cost_time, "for loop", jacrev_cost_time, "jacrev")
```

жӯӨеӨ–пјҢд№ҹеҸҜд»ҘйҖҡиҝҮжҢҮе®ҡ`jacrev`зҡ„еҸӮж•°`grad_position`жқҘи®Ўз®—иҫ“еҮәзӣёеҜ№дәҺжЁЎеһӢеҸӮж•°weightе’Ңbiasзҡ„йӣ…еҸҜжҜ”зҹ©йҳөгҖӮ

```python
jacrev_weight, jacrev_bias = jacrev(forecast, grad_position=(0, 1))(weight, bias, x)
print(jacrev_weight.shape)
print(jacrev_bias.shape)
```

## еҸҚеҗ‘жЁЎејҸи®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳө vs еүҚеҗ‘жЁЎејҸи®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳө

MindSporeжҸҗдҫӣдәҶдёӨдёӘAPIжқҘи®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳөпјҡеҲҶеҲ«жҳҜ`jacrev`е’Ң`jacfwd`гҖӮ

* `jacrev`пјҡдҪҝз”ЁеҸҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶгҖӮ

* `jacfwd`пјҡдҪҝз”ЁеүҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶгҖӮ

`jacfwd`е’Ң`jacrev`еҸҜд»Ҙзӣёдә’жӣҝжҚўпјҢдҪҶжҳҜе®ғд»¬еңЁдёҚеҗҢзҡ„еңәжҷҜдёӢпјҢжҖ§иғҪиЎЁзҺ°дёҚеҗҢгҖӮ

дёҖиҲ¬жқҘиҜҙпјҢеҰӮжһңйңҖиҰҒи®Ўз®—еҮҪж•°$R^{n}\overset{}{\rightarrow}R^{m}$зҡ„йӣ…еҸҜжҜ”зҹ©йҳөпјҢеҪ“иҜҘеҮҪж•°зҡ„иҫ“еҮәеҗ‘йҮҸзҡ„и§„жЁЎеӨ§дәҺиҫ“е…Ҙеҗ‘йҮҸзҡ„и§„жЁЎж—¶пјҲеҚіпјҢm > nпјүпјҢ`jacfwd`еңЁжҖ§иғҪж–№йқўиЎЁзҺ°еҫ—жӣҙеҘҪпјҢеҗҰеҲҷпјҢ`jacrev`еңЁжҖ§иғҪж–№йқўиЎЁзҺ°еҫ—жӣҙеҘҪгҖӮ

дёӢйқўеҜ№иҝҷдёӘз»“и®әеҒҡдёҖдёӘйқһдёҘи°Ёзҡ„и®әиҜҒпјҢеңЁеүҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶпјҲи®Ўз®—йӣ…еҸҜжҜ”-еҗ‘йҮҸз§Ҝпјүзҡ„иҝҮзЁӢдёӯпјҢжҳҜйҖҗеҲ—и®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳөзҡ„пјҢеңЁеҸҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶпјҲи®Ўз®—еҗ‘йҮҸ-йӣ…еҸҜжҜ”з§Ҝпјүзҡ„иҝҮзЁӢдёӯпјҢжҳҜйҖҗиЎҢи®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳөзҡ„гҖӮеҒҮи®ҫеҫ…и®Ўз®—зҡ„йӣ…еҸҜжҜ”зҹ©йҳөзҡ„и§„жЁЎжҳҜmиЎҢпјҢnеҲ—пјҢеҰӮжһңm > nпјҢжҲ‘д»¬жҺЁиҚҗдҪҝз”ЁйҖҗеҲ—и®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳөзҡ„`jacfwd`пјҢеҸҚд№ӢпјҢеҰӮжһңm < nпјҢжҲ‘д»¬жҺЁиҚҗдҪҝз”ЁйҖҗиЎҢи®Ўз®—йӣ…еҸҜжҜ”зҹ©йҳөзҡ„`jacrev`гҖӮ

## й»‘еЎһзҹ©йҳө

еңЁд»Ӣз»ҚMindSporeжҸҗдҫӣзҡ„и®Ўз®—й»‘еЎһзҹ©йҳөзҡ„ж–№жі•д№ӢеүҚпјҢйҰ–е…ҲеҜ№й»‘еЎһзҹ©йҳөиҝӣиЎҢд»Ӣз»ҚгҖӮ

й»‘еЎһзҹ©йҳөеҸҜд»Ҙз”ұжўҜеәҰж“ҚдҪң$\nabla$е’Ңе№ҝеәҰжўҜеәҰж“ҚдҪң$\partial$зҡ„еӨҚеҗҲеҫ—еҲ°пјҢеҚі

$$\nabla \circ \partial: F_{n}^{1} \longrightarrow F_{n}^{n} \longrightarrow F_{n \times n}^{n}$$

е°ҶиҜҘеӨҚеҗҲж“ҚдҪңз”ЁдәҺfпјҢеҫ—еҲ°пјҢ

$$f \longmapsto \nabla f \longmapsto J_{\nabla f}$$

еҸҜд»Ҙеҫ—еҲ°й»‘еЎһзҹ©йҳөпјҢ

$$H_{f} = \begin{bmatrix}
  \frac{\partial (\nabla _{1}f)}{\partial x_{1}} &\frac{\partial (\nabla _{1}f)}{\partial x_{2}} &\dots &\frac{\partial (\nabla _{1}f)}{\partial x_{n}}  \\
  \frac{\partial (\nabla _{2}f)}{\partial x_{1}} &\frac{\partial (\nabla _{2}f)}{\partial x_{2}} &\dots &\frac{\partial (\nabla _{2}f)}{\partial x_{n}} \\
  \vdots &\vdots &\ddots &\vdots \\
  \frac{\partial (\nabla _{n}f)}{\partial x_{1}} &\frac{\partial (\nabla _{n}f)}{\partial x_{2}} &\dots &\frac{\partial (\nabla _{n}f)}{\partial x_{n}}
\end{bmatrix} = \begin{bmatrix}
  \frac{\partial ^2 f}{\partial x_{1}^{2}} &\frac{\partial ^2 f}{\partial x_{2} \partial x_{1}} &\dots &\frac{\partial ^2 f}{\partial x_{n} \partial x_{1}}  \\
  \frac{\partial ^2 f}{\partial x_{1} \partial x_{2}} &\frac{\partial ^2 f}{\partial x_{2}^{2}} &\dots &\frac{\partial ^2 f}{\partial x_{n} \partial x_{2}} \\
  \vdots &\vdots &\ddots &\vdots \\
  \frac{\partial ^2 f}{\partial x_{1} \partial x_{n}} &\frac{\partial ^2 f}{\partial x_{2} \partial x_{n}} &\dots &\frac{\partial ^2 f}{\partial x_{n}^{2}}
\end{bmatrix}$$

жҳ“и§ҒпјҢй»‘еЎһзҹ©йҳөжҳҜдёҖдёӘе®һеҜ№з§°зҹ©йҳөгҖӮ

й»‘еЎһзҹ©йҳөзҡ„еә”з”ЁпјҡеҲ©з”Ёй»‘еЎһзҹ©йҳөпјҢжҲ‘д»¬еҸҜд»ҘжҺўзҙўзҘһз»ҸзҪ‘з»ңеңЁжҹҗзӮ№еӨ„зҡ„жӣІзҺҮпјҢдёәи®ӯз»ғжҳҜеҗҰж”¶ж•ӣжҸҗдҫӣж•°еҖјдҫқжҚ®гҖӮ

## и®Ўз®—й»‘еЎһзҹ©йҳө

еңЁMindSporeдёӯпјҢжҲ‘д»¬еҸҜд»ҘйҖҡиҝҮ`jacfwd`е’Ң`jacrev`зҡ„д»»ж„Ҹз»„еҗҲжқҘи®Ўз®—й»‘еЎһзҹ©йҳөгҖӮ

```python

Din = 32
Dout = 16
weight = ops.randn(Dout, Din)
bias = ops.randn(Dout)
x = ops.randn(Din)

hess1 = jacfwd(jacfwd(forecast, grad_position=2), grad_position=2)(weight, bias, x)
hess2 = jacfwd(jacrev(forecast, grad_position=2), grad_position=2)(weight, bias, x)
hess3 = jacrev(jacfwd(forecast, grad_position=2), grad_position=2)(weight, bias, x)
hess4 = jacrev(jacrev(forecast, grad_position=2), grad_position=2)(weight, bias, x)

np.allclose(hess1.asnumpy(), hess2.asnumpy())
np.allclose(hess2.asnumpy(), hess3.asnumpy())
np.allclose(hess3.asnumpy(), hess4.asnumpy())
```

## и®Ўз®—жү№йҮҸйӣ…еҸҜжҜ”зҹ©йҳөе’Ңжү№йҮҸй»‘еЎһзҹ©йҳө

еңЁдёҠйқўз»ҷеҮәзҡ„зӨәдҫӢдёӯпјҢжҲ‘д»¬йғҪжҳҜи®Ўз®—еҚ•дёҖзҡ„иҫ“еҮәеҗ‘йҮҸе…ідәҺеҚ•дёҖзҡ„иҫ“е…Ҙеҗ‘йҮҸзҡ„йӣ…еҸҜжҜ”зҹ©йҳөгҖӮеңЁжҹҗдәӣжғ…еҶөдёӢпјҢдҪ еҸҜиғҪжғіи®Ўз®—дёҖжү№йҮҸзҡ„иҫ“еҮәеҗ‘йҮҸе…ідәҺдёҖжү№йҮҸзҡ„иҫ“е…Ҙеҗ‘йҮҸзҡ„йӣ…еҸҜжҜ”зҹ©йҳөпјҢжҲ–иҖ…жҚўеҸҘиҜқиҜҙпјҢз»ҷеҮәдёҖжү№йҮҸзҡ„иҫ“е…Ҙеҗ‘йҮҸпјҢе…¶shapeдёә(b, n)пјҢеҮҪж•°зҡ„жҳ е°„е…ізі»жҳҜ$R^{n}\overset{}{\rightarrow}R^{m}$пјҢжҲ‘д»¬жңҹжңӣеҫ—еҲ°дёҖжү№йҮҸзҡ„йӣ…еҸҜжҜ”зҹ©йҳөпјҢе…¶shapeдёә(b, m, n)гҖӮ

жҲ‘д»¬еҸҜд»ҘдҪҝз”Ё`vmap`и®Ўз®—жү№йҮҸйӣ…еҸҜжҜ”зҹ©йҳөгҖӮ

```python
batch_size = 64
Din = 31
Dout = 33

weight = ops.randn(Dout, Din)
bias = ops.randn(Dout)
x = ops.randn(batch_size, Din)

compute_batch_jacobian = vmap(jacrev(forecast, grad_position=2), in_axes=(None, None, 0))
batch_jacobian = compute_batch_jacobian(weight, bias, x)
print(batch_jacobian.shape)
```

и®Ўз®—жү№йҮҸй»‘еЎһзҹ©йҳөзҡ„ж–№ејҸдёҺи®Ўз®—жү№йҮҸйӣ…еҸҜжҜ”зҹ©йҳөзҡ„ж–№жі•зұ»дјјпјҢд№ҹеҸҜд»ҘдҪҝз”Ё`vmap`жқҘи®Ўз®—жү№йҮҸй»‘еЎһзҹ©йҳөпјҢ

```python
hessian = jacrev(jacrev(forecast, grad_position=2), grad_position=2)
compute_batch_hessian = vmap(hessian, in_axes=(None, None, 0))
batch_hessian = compute_batch_hessian(weight, bias, x)
print(batch_hessian.shape)
```

## и®Ўз®—й»‘еЎһ-еҗ‘йҮҸз§Ҝ

и®Ўз®—й»‘еЎһ-еҗ‘йҮҸз§ҜпјҲHessian-vector product, hvpпјүзҡ„жңҖзӣҙжҺҘзҡ„ж–№жі•и®Ўз®—дёҖдёӘе®Ңж•ҙзҡ„й»‘еЎһзҹ©йҳөпјҢе№¶е°Ҷе…¶дёҺеҗ‘йҮҸиҝӣиЎҢзӮ№з§Ҝиҝҗз®—гҖӮдҪҶMindSporeжҸҗдҫӣдәҶжӣҙеҘҪзҡ„ж–№жі•пјҢдҪҝеҫ—дёҚйңҖиҰҒи®Ўз®—дёҖдёӘе®Ңж•ҙзҡ„й»‘еЎһзҹ©йҳөпјҢдҫҝеҸҜд»Ҙи®Ўз®—й»‘еЎһ-еҗ‘йҮҸз§ҜгҖӮдёӢйқўжҲ‘д»¬д»Ӣз»Қи®Ўз®—й»‘еЎһ-еҗ‘йҮҸз§Ҝзҡ„дёӨз§Қж–№жі•гҖӮ

* е°ҶеҸҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶдёҺеҸҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶз»„еҗҲгҖӮ

* е°ҶеҸҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶдёҺеүҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶз»„еҗҲгҖӮ

дёӢйқўе…Ҳд»Ӣз»ҚпјҢеңЁMindSporeдёӯпјҢеҰӮдҪ•дҪҝз”ЁеҸҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶдёҺеүҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶз»„еҗҲзҡ„ж–№ејҸи®Ўз®—й»‘еЎһ-еҗ‘йҮҸз§ҜпјҢ

```python
def hvp_revfwd(f, inputs, vector):
    return jvp(grad(f), inputs, vector)[1]

def f(x):
    return x.sin().sum()

inputs = ops.randn(128)
vector = ops.randn(128)

result_hvp_revfwd = hvp_revfwd(f, inputs, vector)
print(result_hvp_revfwd.shape)
```

еҰӮжһңеүҚеҗ‘иҮӘеҠЁеҫ®еҲҶдёҚиғҪж»Ўи¶іиҰҒжұӮпјҢжҲ‘д»¬еҸҜд»ҘдҪҝз”ЁеҸҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶдёҺеҸҚеҗ‘жЁЎејҸиҮӘеҠЁеҫ®еҲҶз»„еҗҲзҡ„ж–№ејҸжқҘи®Ўз®—й»‘еЎһ-еҗ‘йҮҸз§ҜпјҢ

```python
def hvp_revrev(f, inputs, vector):
    _, vjp_fn = vjp(grad(f), *inputs)
    return vjp_fn(*vector)

result_hvp_revrev = hvp_revrev(f, (inputs,), (vector,))
print(result_hvp_revrev[0].shape)
```

дҪҝз”ЁдёҠйқўдёӨз§Қж–№жі•и®Ўз®—й»‘еЎһ-еҗ‘йҮҸз§Ҝеҫ—еҲ°зҡ„з»“жһңжҳҜдёҖж ·зҡ„гҖӮ

```python
assert np.allclose(result_hvp_revfwd.asnumpy(), result_hvp_revrev[0].asnumpy())
```
